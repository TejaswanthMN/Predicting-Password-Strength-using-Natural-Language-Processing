```python
#importing all the necessary modules
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("data.csv",error_bad_lines = False)
data.head()
```

```python
#checking the unique values
data["strength"].unique()
```

```python
#checking null values
data.isna().sum()
```

```python
#removing the null values
data.dropna(inplace = True)
```

```python
data.isna().sum()
```

```python
sns.countplot(data["strength"])
```

```python
#as we could see a large number of values belongs to 1, there is a chance that the machine learning model could be biased about that value
#converting the data frame to a array
password_array = np.array(data)
password_array
```

```python
#shuffling the data to ensure robustness
import random
random.shuffle(password_array)
password_array
```

```python
#extracting independent and dependent data

x = [labels[0] for labels in password_array]
y = [labels[1] for labels in password_array]
```

```python
x
```

```python
y
```

```python
#function for seperating each charcter for vectorizing based on the importance using TF IDF Vectorizer
def word_divide_char(st):
    characters = []
    for i in st:
        characters.append(i)
    return characters
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer(tokenizer = word_divide_char)
X = vect.fit_transform(x)
X
```

```python
X.shape
```

```python
#to see all the features
vect.get_feature_names()
```

```python
first_document_vector = X[0]
first_document_vector.T.todense()
```

```python
#converting to a dataframe with index as character and the value as importance
df = pd.DataFrame(first_document_vector.T.todense(),index=vect.get_feature_names(),columns = ['TF-IDF'])
df.sort_values("TF-IDF",ascending = False)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

```python
X_train.shape
```

```python
#having more than  3 classes (0, 1, 2) so multi class is sselecting as  multinomial
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state = 0, multi_class = 'multinomial')
clf.fit(X_train, y_train)
```

```python
#predicitng the strength of some random passsword
dt = np.array(["%@a1b2c3d4"])
pred = vect.transform(dt)
clf.predict(pred)
```

```python
#predicting the test data
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```

```python
#importing classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
